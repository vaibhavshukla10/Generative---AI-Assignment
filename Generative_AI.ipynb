2. Explain the role of probabilistic modeling in generative models. How do
these models differ from discriminative models?

      - Role of Probabilistic Modeling in Generative Models:

           Probabilistic modeling is the foundation of generative models. It enables the model to learn the probability distribution of the data, allowing it to generate new samples that resemble the training data. Instead of memorizing examples, the model estimates how likely different data patterns are and then creates new data based on those probabilities.

           For example, a generative model trained on handwritten digits can generate new digit images that look realistic but were never seen during training.

           Key Roles:

           - Learns the underlying data distribution.
           - Generates new and realistic data samples.
           - Handles uncertainty and variability in data.
           - Supports tasks such as image generation, text generation, speech synthesis, and data augmentation.


          # Difference Between Generative and Discriminative Models

          # Generative Models

          - Learn the joint probability distribution P(X, Y) or data distribution P(X).
          - Can generate new data samples.
          - Focus on how the data is generated.
          - Used for content generation and unsupervised learning.
          - Examples: GANs, VAEs, GPT, Diffusion Models.


          # Discriminative Models

          - Learn the conditional probability P(Y|X).
          - Cannot generate new data; only classify or predict.
          - Focus on distinguishing between classes.
          - Used mainly for classification and prediction tasks.
          - Examples: Logistic Regression, Support Vector Machine (SVM), Decision Trees, Neural Network Classifiers.




          Conclusion - Probabilistic modeling allows generative models to understand the underlying data distribution and create realistic new content. In contrast, discriminative models focus on learning the relationship between inputs and outputs to make accurate predictions or classifications, without generating new data.

3.  What is the difference between Autoencoders and Variational
Autoencoders (VAEs) in the context of text generation?

    # Autoencoders (AEs)

      - Compress input text into a fixed latent representation and reconstruct it.
      - Produce a deterministic latent vector.
      - Mainly used for text compression, denoising, and feature extraction.
      - Cannot easily generate diverse new text.
      - Simpler architecture and easier to train.


      # Variational Autoencoders (VAEs)

      - Learn a probability distribution in the latent space and generate new text by sampling from it.
      - Produce a probabilistic latent representation (mean and variance).
      - Mainly used for text generation, sentence completion, and data augmentation.
      - Can generate diverse and realistic text samples.
      - More complex due to probabilistic sampling and additional loss function (KL divergence).

4. Describe the working of attention mechanisms in Neural Machine
Translation (NMT). Why are they critical?

      - Attention Mechanism in Neural Machine Translation (NMT):

           The attention mechanism allows an NMT model to focus on the most relevant words in the source sentence while generating each word in the target sentence. Instead of relying on a single fixed-length context vector, the decoder dynamically attends to different parts of the input sequence at every decoding step.

           Working of Attention
           - Encoding: The encoder processes the input sentence and produces a hidden state for each input word.
           - Attention Score: For each target word, the decoder calculates attention scores between its current state and all encoder hidden states.
           - Context Vector: A weighted sum of the encoder hidden states is computed to form a context vector.
           - Word Generation: The decoder uses the context vector along with its previous state to generate the next target word.
           - Repeat: This process continues until the entire translated sentence is generated


           Why is Attention Critical?
             
          Handles long sentences by avoiding the limitations of a fixed-length context vector.
          Improves translation accuracy by focusing on the most relevant source words.
          Captures long-range dependencies between words in a sentence.

6.  Use the following small text dataset to train a simple Variational
Autoencoder (VAE) for text reconstruction:
["The sky is blue", "The sun is bright", "The grass is green",
"The night is dark", "The stars are shining"]
   
    1.Preprocess the data (tokenize and pad the sequences).
     
   2. Build a basic VAE model for text reconstruction.
   3. Train the model and show how it reconstructs or generates similar sentences.
Include your code, explanation, and sample outputs.
(Include your Python code and output in the code box below.)

      - This is a programming assignment, so below is a complete example using TensorFlow/Keras. Since the dataset is extremely small (5 sentences), the VAE will not generate high-quality text. The purpose is to demonstrate the implementation.


```python
# ===============================
# Variational Autoencoder (VAE)
# for Simple Text Reconstruction
# ===============================

import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.layers import Input, Embedding, LSTM, Dense, Lambda
from tensorflow.keras.models import Model
from tensorflow.keras import backend as K

# -------------------------------
# Step 1: Dataset
# -------------------------------

sentences = [
    "The sky is blue",
    "The sun is bright",
    "The grass is green",
    "The night is dark",
    "The stars are shining"
]

# -------------------------------
# Step 2: Tokenization & Padding
# -------------------------------

tokenizer = Tokenizer()
tokenizer.fit_on_texts(sentences)

sequences = tokenizer.texts_to_sequences(sentences)

max_len = max(len(s) for s in sequences)

X = pad_sequences(sequences, maxlen=max_len, padding='post')

vocab_size = len(tokenizer.word_index) + 1

print("Vocabulary Size:", vocab_size)
print("Padded Sequences:\n", X)

# -------------------------------
# Step 3: Build VAE
# -------------------------------

latent_dim = 16
embedding_dim = 32

inputs = Input(shape=(max_len,))

embedding = Embedding(vocab_size, embedding_dim)(inputs)

encoded = LSTM(32)(embedding)

z_mean = Dense(latent_dim)(encoded)
z_log_var = Dense(latent_dim)(encoded)

def sampling(args):
    z_mean, z_log_var = args
    epsilon = K.random_normal(shape=(K.shape(z_mean)[0], latent_dim))
    return z_mean + K.exp(0.5 * z_log_var) * epsilon

z = Lambda(sampling)([z_mean, z_log_var])

decoder = Dense(max_len * embedding_dim, activation="relu")(z)
decoder = tf.keras.layers.Reshape((max_len, embedding_dim))(decoder)

decoded = LSTM(32, return_sequences=True)(decoder)

outputs = Dense(vocab_size, activation="softmax")(decoded)

vae = Model(inputs, outputs)

# -------------------------------
# Step 4: VAE Loss
# -------------------------------

reconstruction_loss = tf.keras.losses.sparse_categorical_crossentropy(
    X,
    outputs
)

reconstruction_loss = K.sum(reconstruction_loss)

kl_loss = -0.5 * K.sum(
    1 + z_log_var - K.square(z_mean) - K.exp(z_log_var)
)

vae_loss = reconstruction_loss + kl_loss

vae.add_loss(vae_loss)

vae.compile(optimizer="adam")

# -------------------------------
# Step 5: Train
# -------------------------------

vae.fit(X, epochs=100, batch_size=2)

# -------------------------------
# Step 6: Reconstruction
# -------------------------------

pred = vae.predict(X)

pred_words = np.argmax(pred, axis=-1)

reverse_word_index = {v:k for k,v in tokenizer.word_index.items()}

print("\nReconstructed Sentences:\n")

for seq in pred_words:
    sentence = []
    for idx in seq:
        if idx != 0:
            sentence.append(reverse_word_index.get(idx,""))
    print(" ".join(sentence))
```

```python
# ==========================================
# GPT-2 Translation (English → French/German)
# ==========================================

from transformers import pipeline

# Load GPT-2 text generation pipeline
generator = pipeline("text-generation", model="gpt2")

# Original English paragraph
text = "Artificial Intelligence is transforming the world by improving healthcare, education, and business."

# ---------- English to French ----------
prompt_fr = f"Translate the following English text to French:\n\n{text}\n\nFrench:"
result_fr = generator(
    prompt_fr,
    max_new_tokens=60,
    do_sample=False
)

# ---------- English to German ----------
prompt_de = f"Translate the following English text to German:\n\n{text}\n\nGerman:"
result_de = generator(
    prompt_de,
    max_new_tokens=60,
    do_sample=False
)

print("Original:")
print(text)

print("\nFrench Translation:")
print(result_fr[0]["generated_text"])

print("\nGerman Translation:")
print(result_de[0]["generated_text"])
```

8.  Implement a simple attention-based encoder-decoder model for
English-to-Spanish translation using Tensorflow or PyTorch.
(Include your Python code and output in the code box below.)
  
       - Below is a simple implementation of an Attention-based Encoder-Decoder (Seq2Seq) model for English-to-Spanish translation using TensorFlow/Keras. This example uses a very small dataset for demonstration purposes.

```python
# ============================================
# Attention-based Encoder-Decoder for
# English to Spanish Translation
# ============================================

import tensorflow as tf
from tensorflow.keras.layers import Input, Embedding, LSTM, Dense, Attention, Concatenate
from tensorflow.keras.models import Model
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
import numpy as np

# -------------------------------
# Sample Dataset
# -------------------------------

english = [
    "hello",
    "how are you",
    "good morning",
    "thank you",
    "good night"
]

spanish = [
    "hola",
    "como estas",
    "buenos dias",
    "gracias",
    "buenas noches"
]

# -------------------------------
# Tokenization
# -------------------------------

eng_tokenizer = Tokenizer(filters='')
spa_tokenizer = Tokenizer(filters='')

eng_tokenizer.fit_on_texts(english)
spa_tokenizer.fit_on_texts(spanish)

encoder_input = eng_tokenizer.texts_to_sequences(english)
decoder_input = spa_tokenizer.texts_to_sequences(spanish)

max_eng = max(len(x) for x in encoder_input)
max_spa = max(len(x) for x in decoder_input)

encoder_input = pad_sequences(encoder_input, maxlen=max_eng, padding='post')
decoder_input = pad_sequences(decoder_input, maxlen=max_spa, padding='post')

encoder_vocab = len(eng_tokenizer.word_index) + 1
decoder_vocab = len(spa_tokenizer.word_index) + 1

# Decoder target
decoder_target = np.expand_dims(decoder_input, -1)

# -------------------------------
# Encoder
# -------------------------------

latent_dim = 64

encoder_inputs = Input(shape=(max_eng,))
enc_embed = Embedding(encoder_vocab, 64)(encoder_inputs)

encoder_outputs, state_h, state_c = LSTM(
    latent_dim,
    return_sequences=True,
    return_state=True
)(enc_embed)

# -------------------------------
# Decoder
# -------------------------------

decoder_inputs = Input(shape=(max_spa,))
dec_embed = Embedding(decoder_vocab, 64)(decoder_inputs)

decoder_outputs, _, _ = LSTM(
    latent_dim,
    return_sequences=True,
    return_state=True
)(
    dec_embed,
    initial_state=[state_h, state_c]
)

# -------------------------------
# Attention Layer
# -------------------------------

attention = Attention()
attention_output = attention([decoder_outputs, encoder_outputs])

concat = Concatenate(axis=-1)(
    [decoder_outputs, attention_output]
)

outputs = Dense(
    decoder_vocab,
    activation='softmax'
)(concat)

# -------------------------------
# Build Model
# -------------------------------

model = Model(
    [encoder_inputs, decoder_inputs],
    outputs
)

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

model.summary()

# -------------------------------
# Train
# -------------------------------

model.fit(
    [encoder_input, decoder_input],
    decoder_target,
    epochs=100,
    batch_size=2
)

# -------------------------------
# Prediction
# -------------------------------

prediction = model.predict(
    [encoder_input, decoder_input]
)

predicted = np.argmax(prediction, axis=-1)

reverse_index = {
    v:k for k,v in spa_tokenizer.word_index.items()
}

print("\nTranslations:\n")

for seq in predicted:
    words = []
    for token in seq:
        if token != 0:
            words.append(reverse_index.get(token, ""))
    print(" ".join(words))
```

Explanation
   - Dataset
     A small English–Spanish parallel corpus is created.
   - Preprocessing
   - Tokenize both English and Spanish sentences.
     Convert words to integer sequences.
     Pad sequences to equal length.
   - Encoder
     Uses an Embedding layer followed by an LSTM.


Note: Since the dataset contains only five sentence pairs, the model memorizes the training data. In real-world Neural Machine Translation, attention-based encoder-decoder models are trained on thousands or millions of bilingual sentence pairs for accurate translation.
